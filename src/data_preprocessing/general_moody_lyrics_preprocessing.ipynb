```python
lyrics = []
```

```python
import pandas as pd
import time
import random

# Define the function to retrieve lyrics
def getLyrics(row):
    
#     wait_time = random.randint(1, 4)  # You can adjust the range as needed
#     print(f"Waiting for {wait_time} seconds...")
#     time.sleep(wait_time)
    
    artist = genius.search_artist(row["Artist"], max_songs=1)
    if artist is None:
        lyrics.append("NO")
        return
    song = artist.song(row["Title"])
    if song is None:
        lyrics.append("NO")
        return
    if song.lyrics is None:
        lyrics.append("NO")
        return
    
    lyrics.append(song.lyrics)
    return

for i in range(moody_lyrics.shape[0]):
    getLyrics(moody_lyrics.iloc[i])
    print("Done with: ", i)

# Now, the 'getLyrics' function correctly accesses the 'artist' and 'title' columns of the DataFrame.
```
