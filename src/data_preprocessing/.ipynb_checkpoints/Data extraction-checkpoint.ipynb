## Read the links for all Taylor Swift's songs

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.azlyrics.com/t/taylorswift.html')
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
divTags = soup.find_all('div', class_= 'listalbum-item')

hrefLinks = [div.a.get('href') for div in divTags]
```

```python
print("Total songs are: " + str(len(hrefLinks)))
```

## Fetch the lyrics of each song using the link

```python
from selenium import webdriver
import time
import pandas as pd
import random
```

```python
baseUrl = 'https://www.azlyrics.com'
```

```python
# Use this if you are running the script for the first time
df = pd.DataFrame(columns=['song_name', 'lyrics'])

# Use this if you are reusing the script
# df = pd.read_csv('../data/partial_data.csv')
```

```python
driver = webdriver.Chrome("/Users/adityasingh/Downloads/chromedriver_mac64/chromedriver")
```

```python
# check for entry 285
for i in range(0,len(hrefLinks)):
    print("Fetching data for lyrics " + str(i) + " and link is: " + hrefLinks[i])
    songLink = baseUrl + hrefLinks[i]
    driver.get(songLink)
    waitTime = random.randrange(60)
    print("Wait time is: " + str(waitTime))
    time.sleep(waitTime)
    pageContent = BeautifulSoup(driver.page_source, 'html.parser')
    previousDivTag = pageContent.find('div', class_='ringtone')
    nameOfSongTag = previousDivTag.find_next_sibling('b')
    nameOfSong = nameOfSongTag.text
    lyricsTag = previousDivTag.find_next_sibling('div')
    lyrics = lyricsTag.text
    df = df.append({'song_name': nameOfSong, 'lyrics': lyrics}, ignore_index=True)
```

```python
driver.quit()
```
