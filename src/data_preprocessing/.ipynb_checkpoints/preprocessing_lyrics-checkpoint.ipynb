```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import re
import requests
from nltk.corpus import stopwords
import datetime
```

```python
data = pd.read_csv('../../data/uncleaned-data-with-features.csv')
```

```python
data
```

```python
def countWords(s):
    split = s.split(' ')
    return len(split)
```

```python
def removeUnneccessaryCharacters(s):
    text = re.sub(r'[^\w\s]|_', '', s)
    text = text.lstrip('\n')
    return text
```

```python
stop_words = set(stopwords.words('english'))
def countStopWords(s):
    count = 0
    for word in s.split(' '):
        if word in stop_words:
            count += 1
    return count
```

```python
def countLines(s):
    lines = s.split('\n')
    ret = []
    for line in lines:
        if len(line) > 0:
            ret.append(line)
            
    return len(ret)
```

```python
stop_words = set(stopwords.words('english'))
def removeStopWords(s):
    s = s.lstrip('\n')
    words = " ".join([word for word in s.split(' ') if word not in stop_words])
    return words
```

```python
def countUniqueWords(s):
    uniqueWords = set(s.split())
    return len(uniqueWords)
```

```python
def getDate(s):
    x = datetime(s)
    print(x)
```

```python
data['lyrics'] = data.lyrics.apply(removeUnneccessaryCharacters)
data['stop_word_count'] = data.lyrics.apply(countStopWords)
data['word_count'] = data.lyrics.apply(countWords)
data['line_count'] = data.lyrics.apply(countLines)
data['avg_words_per_line'] = data.word_count / data.line_count
data['cleaned_lyrics'] = data.lyrics.apply(removeStopWords)
data['song_name'] = data.song_name.apply(lambda x:  x.replace('\"',''))
data['number_of_unique_words'] = data.lyrics.apply(countUniqueWords)
```

```python
data.release_data = pd.to_datetime(data['release_data'], format='%Y-%m-%d')
```

```python
data['month'] = data.release_data.dt.month
data['year'] = data.release_data.dt.year
data['day'] = data.release_data.dt.day
```

```python
data
```

```python
data.to_pickle('../../data/cleaned-data-with-features_date.pkl')
data.to_csv('../../data/cleaned-data-with-features_date.csv')
```

## Linear regression

```python
normalData.columns
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
columns = ['stop_word_count', 'word_count', 'line_count','words_per_line', 
           'number_of_unique_words', 'playcount']
```

```python
X = normalData[columns]
y = normalData.rating
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
lin = LinearRegression()
```

```python
lin.fit(X_train,y_train)
```

```python
y_predict_lin = lin.predict(X_test)
```

```python
r2_score(y_test, y_predict_lin)
```

```python
rfr = RandomForestRegressor()
```

```python
rfr.fit(X_train, y_train)
```

```python
y_predict_rfr = rfr.predict(X_test)
```

```python
r2_score(y_test, y_predict_rfr)
```
